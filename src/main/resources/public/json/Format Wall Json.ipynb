```python
def comp (a, b):
    if (a[0] == b[0] and a[1] > b[1]):
        return True
    elif (a[1] == b[1] and a[0] > b[0]):
        return True
    return False
```

```python
import json
with open('./wall 2.json') as f:
  data = json.load(f)
for i in data['wall']:
    if (comp(i["start"], i["end"])):
        temp = i["start"]
        i["start"] = i["end"]
        i["end"] = temp
with open('wall 2 new.json', 'w') as json_file:
  json.dump(data, json_file, indent=2, separators=(',', ': '))
print(json.dumps(data, indent=2, separators=(',', ': ')))
```
