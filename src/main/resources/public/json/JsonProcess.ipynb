```python
import json
with open('./wall copy.json') as f:
  data = json.load(f)
for i in data['wall']: 
    i["start"][0] = int(i["start"][0]*1.5)
    i["start"][1] = int(i["start"][1]*1.5)
    i["end"][0] = int(i["end"][0]*1.5)
    i["end"][1] = int(i["end"][1]*1.5)
with open('wall new.json', 'w') as json_file:
  json.dump(data, json_file, indent=2, separators=(',', ': '))
print(json.dumps(data, indent=2, separators=(',', ': ')))
```
